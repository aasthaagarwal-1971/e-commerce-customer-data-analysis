```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#getting the data
customers = pd.read_csv("Ecommerce Customers")
```

```python
#starting with exploratory data analysis
sns.set_palette("GnBu_d")
sns.set_style('whitegrid')
```

```python
#the data looks fairly distributed with hours being high=37
sns.jointplot(x='Time on Website',y='Yearly Amount Spent',data=customers)
```

```python
#plot is fairly distributed at time=12
sns.jointplot(x='Time on App',y='Yearly Amount Spent',data=customers)
```

```python
sns.jointplot(x='Time on App',y='Length of Membership',kind='hex',data=customers)
```

```python
#to get a better picture we use pairplot to correlate all the parameters
sns.pairplot(customers)
```

```python
#Based off this plot length pf membership looks to be the most correlated feature with Yearly Amount Spent
```

```python
#so we create a model to fit in this data, linear model
sns.lmplot(x='Length of Membership',y='Yearly Amount Spent',data=customers)
```

```python
#training and testing data
X = customers[['Avg. Session Length', 'Time on App','Time on Website', 'Length of Membership']]
y = customers['Yearly Amount Spent']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
#fiting in the model
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
#the coefficient tells us about how each feature is related to each other
print('coefficients:', lm.coef_)
```

```python
#predicting the test data
predictions = lm.predict(X_test)
plt.scatter(y_test,predictions)
```

```python
#evaluating the model
from sklearn import metrics
print('MAE=',metrics.mean_absolute_error(y_test,predictions))
print('MSE=',metrics.mean_squared_error(y_test,predictions))
print('RMSE=',np.sqrt(metrics.mean_squared_error(y_test,predictions)))
```

```python
sns.distplot(y_test-predictions)
```

```python
#the plot is distributed, this implies that the model was a good fit
```

```python
#now we want to see does it even matter to be on website or app for the company
coeffecients = pd.DataFrame(lm.coef_,X.columns)
coeffecients.columns = ['Coeffecient']
coeffecients
```

```python
#INTERPRETATIONS MADE
#Each coefficient represents the expected increase in the target variable (e.g., yearly amount spent) for a one-unit increase in the corresponding feature, keeping all other variables constant.
#Time on app has a strong positive impact, suggesting that increased engagement on the mobile app significantly contributes to higher customer spending.
#Length of customers is the highest coefficient, implying that long-term customers are the most valuable, and retention plays a critical role in revenue generation.

```

```python
#CONCLUSION
#Based on the model coefficients, the company should focus more on improving the mobile app rather than the website. Time spent on the app has a significantly higher impact on customer spending compared to time spent on the website. Additionally, strategies aimed at increasing customer retention (length of membership) should be prioritized, as it has the strongest influence on revenue.
```
